```python
import json
import requests
import subprocess
import os

print(os.getcwd())
process = subprocess.Popen(
    ["./gradlew", "server:startServer"],
    cwd=".",
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True,
    bufsize=1  # line-buffered
)

for line in process.stdout:
    print(line, end="")
    if "Server started" in line:
        print("✅ Server is ready!")
        break
```

```python
def printErrors(response):
    print("Errors:")
    for error in response["errors"]:
        print(f"\t{error}")

model = """
/*
 * Speedometer model in CRML
 *
 * Models a vehicle speedometer with:
 *   - Speed input (from a wheel sensor, in km/h)
 *   - Speed zone classification (stopped / normal / warning / danger)
 *   - Requirements on display accuracy and speed-limit compliance
 *   - An overspeed alert period tracked over time
 */

model Speedometer is {

    // -------------------------------------------------------------------------
    // Physical unit definitions
    // -------------------------------------------------------------------------

    /* Velocity expressed in km/h.
       SI unit is m/s; 1 km/h = 1/3.6 m/s, so rate = 1/3.6, offset = 0. */
    partial type Quantity is (Real q is rate * u + offset) {
        String SIUnit;
        Real   u;
        Real   rate;
        Real   offset;
    };

    partial type Velocity is Quantity (SIUnit = "m/s");
    type VelocityKmh is Velocity (u is external, rate = 0.27778, offset = 0) alias kmh;

    // -------------------------------------------------------------------------
    // Constants
    // -------------------------------------------------------------------------

    constant Real warningThreshold_kmh is 90.0;   // km/h — speed-warning zone begins
    constant Real dangerThreshold_kmh  is 130.0;  // km/h — legal motorway limit (danger zone)
    constant Real maxDisplayed_kmh     is 260.0;  // km/h — maximum displayable speed
    constant Real accuracyTolerance    is 2.0;    // km/h — permitted display error

    // -------------------------------------------------------------------------
    // Inputs (provided by the vehicle behavioral model)
    // -------------------------------------------------------------------------

    /* Actual vehicle speed as measured by the wheel sensor */
    VelocityKmh actualSpeed is external;

    /* Speed value currently shown on the dial */
    VelocityKmh displayedSpeed is external;

    /* True when the vehicle ignition is on */
    Boolean ignitionOn is external;

    // -------------------------------------------------------------------------
    // Derived signals — speed zones
    // -------------------------------------------------------------------------

    /* Vehicle is considered stopped below 1 km/h */
    Boolean isStopped is actualSpeed < 1.0;

    /* Normal driving: above stopped, below warning threshold */
    Boolean isNormal is actualSpeed >= 1.0 and actualSpeed < warningThreshold_kmh;

    /* Warning zone: approaching the legal limit */
    Boolean isWarning is actualSpeed >= warningThreshold_kmh and actualSpeed < dangerThreshold_kmh;

    /* Danger zone: legal limit exceeded */
    Boolean isDanger is actualSpeed >= dangerThreshold_kmh;

    /* Display is within the physical scale of the dial */
    Boolean isInRange is displayedSpeed >= 0.0 and displayedSpeed <= maxDisplayed_kmh;

    /* Absolute error between actual and displayed speed */
    Real displayError is if actualSpeed >= displayedSpeed
                         then actualSpeed - displayedSpeed
                         else displayedSpeed - actualSpeed;

    Boolean displayAccurate is displayError <= accuracyTolerance;

    // -------------------------------------------------------------------------
    // Time periods
    // -------------------------------------------------------------------------

    /* Operator: open a period while a Boolean is true, close it when it turns false.
       This is the standard FORM-L "during" pattern. */
    Operator [ Periods ] during Boolean b = Periods [ new Clock b, new Clock not b ];

    /* The session lasts as long as the ignition is on */
    Periods sessionActive is during ignitionOn;

    /* Continuous overspeed periods */
    Periods overspeedPeriod is during isDanger;

    /* Warning-zone periods */
    Periods warningPeriod is during isWarning;

    // -------------------------------------------------------------------------
    // Requirements
    // -------------------------------------------------------------------------

    /* Type Requirement is a Boolean reserved for requirement expressions,
       preventing accidental use of temporal accumulation operators on them. */
    type Requirement is Boolean forbid { *, +, integrate };

    /*
     * R1 — Display in range
     * The displayed speed must always lie within the physical scale of the dial
     * while the ignition is on.
     */
    Requirement r_displayInRange is during ignitionOn ensure isInRange;

    /*
     * R2 — Display accuracy
     * While the vehicle is moving, the displayed speed must be within
     * accuracyTolerance km/h of the actual speed.
     */
    Requirement r_displayAccuracy is during isNormal ensure displayAccurate;

    /*
     * R3 — Accuracy maintained in warning zone
     * Accuracy is also required in the warning zone, where reading errors
     * are more consequential.
     */
    Requirement r_warningAccuracy is during isWarning ensure displayAccurate;

    /*
     * R4 — Overspeed must be transient
     * The vehicle must not remain in the danger zone for more than 10 seconds
     * continuously.  Expressed as: check that each overspeed period lasts < 10 s.
     *
     * Uses the FORM-L "check … over" pattern:
     *   check <condition> over <Periods>
     */
    Requirement r_overspeedDuration is check (duration isDanger on overspeedPeriod) < 10.0
                                       over overspeedPeriod;

    /*
     * R5 — Display frozen when stopped
     * When the vehicle is stopped the displayed speed must read zero (within tolerance).
     */
    Requirement r_stoppedDisplay is during isStopped ensure displayedSpeed <= accuracyTolerance;

};


"""

# Send POST request
try:
    response = requests.post(
        "http://localhost:63028/crml/syntax",
        json={"model": model}, 
        timeout=30 # second
    )

    print("Status code:", response.status_code)
    printErrors(json.loads(response.text))

except requests.exceptions.RequestException as e:
    print("Error sending request:", e)
```
